```python
from sklearn.feature_extraction.text import CountVectorizer
# Example documents (LLM-related sentences)
documents = [
"GPT models can generate text and answer questions",
"BERT is used for understanding context in sentences",
"LLAMA models are efficient for large-scale tasks"
]
# Create Bag of Words model
vectorizer = CountVectorizer()
# Fit the model and transform the documents
bow_matrix = vectorizer.fit_transform(documents)
# Convert sparse matrix to a normal array
bow_array = bow_matrix.toarray()
# Print vocabulary and BoW vectors
print("Vocabulary:", vectorizer.get_feature_names_out())
print("BoW Vectors:\n", bow_array)
```
