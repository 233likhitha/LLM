```python
# Step 1: Install required libraries
!pip install transformers torch
# Step 2: Import required modules
from transformers import pipeline
# Step 3: Load sentiment analysis pipeline
sentiment_analyzer = pipeline(     "sentiment-analysis",
model="distilbert-base-uncased-finetuned-sst-2-english"
)
# Step 4: Single sentence analysis
text = "This AI service exceeded my expectations and was very helpful."
single_result = sentiment_analyzer(text)
print("Single Sentence Result:")
print(single_result)
# Step 5: Multiple sentences analysis
texts = [
    "AI makes work easier.",
    "This AI tool is terrible.",
    "The AI system is okay.",
    "Amazing AI performance."
]
multiple_results = sentiment_analyzer(texts)
print("\nMultiple Sentences Results:")
for text, result in zip(texts, multiple_results):
    print(f"Text: {text}")
    print(f"Sentiment: {result['label']}")
    print(f"Confidence Score: {result['score']:.4f}")
    print("-" * 50)
# Step 6: User input sentiment analysis
user_text = input("\nEnter a sentence for sentiment analysis: ")
user_result = sentiment_analyzer(user_text)
print("\nUser Input Result:")
print(user_result)
```
