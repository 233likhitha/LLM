```python
#4A.Named Entity Recognition
# Install transformers
!pip install transformers
# Import required modules
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
# Load a pre-trained NER model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
# Create a Named Entity Recognition pipeline
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)
# Sample text related to Deep Learning
text = "Deep learning models like BERT and GPT are developed by OpenAI and Google for AI research."
# Apply NER
ner_results = ner_pipeline(text)
# Print results
print("Named Entities using Hugging Face Transformers:")
for entity in ner_results:
 print(f"Entity: {entity['word']} | Type: {entity['entity']}")
```

```python
#4B.Question answering

# Simple Question Answering based Text Classification
class SimpleQASystem:
    def __init__(self):
        # Knowledge base
        self.knowledge_base = {
            "what is deep learning?": "Deep learning is a subset of machine learning that uses neural networks with multiple layers.",
            "what is bert?": "BERT is a transformer-based deep learning model used for natural language processing tasks."
        }

    def preprocess(self, question):
        # Convert to lowercase for matching
        return question.strip().lower()

    def get_answer(self, question):
        processed_question = self.preprocess(question)
        if processed_question in self.knowledge_base:
            return self.knowledge_base[processed_question]
        else:
            return "Answer not found"

# Create system
qa_system = SimpleQASystem()

# Given Questions
questions = [
    "What is Deep Learning?",
    "What is BERT?"
]

# Get Answers
for q in questions:
    answer = qa_system.get_answer(q)
    print("Question:", q)
    print("Answer:", answer)
    print("-" * 40)
```

```python
#4C.Text summarization

# Install required libraries
!pip install transformers datasets --quiet
from transformers import BartTokenizer, BartForConditionalGeneration
# Load BART tokenizer and model
tokenizer = BartTokenizer.from_pretrained("facebook/bart-base")
model = BartForConditionalGeneration.from_pretrained("facebook/bart-base")
# Deep Learning–related input text
text = """
Deep learning is a subset of machine learning that uses neural networks with many layers.
Models such as CNNs, RNNs, BERT, and GPT have achieved state-of-the-art performance in tasks
like image recognition, natural language processing, and speech recognition.
These models require large datasets and powerful hardware such as GPUs to train efficiently.
"""
print("Original Text:\n", text, "\n")
# Tokenize input
inputs = tokenizer([text], max_length=512, return_tensors="pt", truncation=True)
# Generate summary
summary_ids = model.generate(
inputs["input_ids"],
max_length=50,
min_length=20,
length_penalty=2.0,
num_beams=4,
early_stopping=True
)
# Decode summary
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print("Generated Summary:\n", summary)
```
